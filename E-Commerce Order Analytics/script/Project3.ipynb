```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('ecommerce.db')

df = pd.read_excel(r"C:\Users\PREM\OneDrive\Desktop\Internship_Project\processed\cleaned_dataset.xlsx")

df.to_sql('orders', conn, if_exists='replace', index=False)

cursor = conn.cursor()
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
print(cursor.fetchall())
```

```python
pd.read_sql_query("SELECT * FROM orders LIMIT 5", conn)
```

```python
pd.read_sql_query("SELECT OrderID, Product, Quantity, TotalPrice FROM orders LIMIT 10", conn)
```

```python
pd.read_sql_query("SELECT * FROM orders WHERE OrderStatus = 'Delivered'", conn)
```

```python
pd.read_sql_query("SELECT * FROM orders WHERE TotalPrice > 2000", conn)
```

```python
pd.read_sql_query("SELECT * FROM orders WHERE Product LIKE 'L%'", conn)
```

```python
pd.read_sql_query("SELECT OrderID, Product, TotalPrice FROM orders ORDER BY TotalPrice DESC LIMIT 10", conn)
```

```python
pd.read_sql_query("""
SELECT Product, COUNT(*) AS OrderCount
FROM orders
GROUP BY Product
ORDER BY OrderCount DESC
""", conn)
```

```python
pd.read_sql_query("""
SELECT Product, SUM(TotalPrice) AS TotalRevenue
FROM orders
GROUP BY Product
ORDER BY TotalRevenue DESC
""", conn)
```

```python
pd.read_sql_query("""
SELECT PaymentMethod, AVG(TotalPrice) AS AvgOrderValue
FROM orders
GROUP BY PaymentMethod
ORDER BY AvgOrderValue DESC
""", conn)
```

```python
pd.read_sql_query("""
SELECT OrderStatus,
       COUNT(*) AS TotalOrders,
       SUM(TotalPrice) AS Revenue,
       AVG(TotalPrice) AS AvgValue
FROM orders
GROUP BY OrderStatus
ORDER BY Revenue DESC
""", conn)
```

```python
pd.read_sql_query("""
SELECT ReferralSource, COUNT(*) AS Orders
FROM orders
WHERE OrderStatus = 'Delivered'
GROUP BY ReferralSource
ORDER BY Orders DESC
""", conn)
```
